## Selecting Five Countries


```python
def countries (row): 
    if row['IDCNTRY'] == 152: 
        return 'Chile'
    if row['IDCNTRY'] == 250: 
        return 'France'
    if row['IDCNTRY'] == 926:
        return 'England'
    if row['IDCNTRY'] == 392:
        return 'Japan'
    if row['IDCNTRY'] == 792:
        return 'Turkey'
    
```

```python
df_timss_students['country'] = df_timss_students.apply(lambda row : countries(row), axis = 1)
```

'University or Higher': 5
'Post-secondary but not University': 4
'Upper Secondary': 3
'Lower Secondary': 2
'Some Primary, Lower Secondary or No School':1 

```python
def parent_edu(row):
    if row['parent_edu_max'] == 'University or Higher':
        return 5
    if row['parent_edu_max'] == 'Post-secondary but not University':
        return 4
    if row['parent_edu_max'] == 'Upper Secondary': 
        return 3
    if row['parent_edu_max'] == 'Lower Secondary':
        return 2
    if row['parent_edu_max'] == 'Some Primary, Lower Secondary or No School':
        return 1
    
```

```python
df_fil_scores['qparent_edu'] = df_fil_scores.apply(lambda row : parent_edu(row), axis = 1)
```

```python
df_fil_scores
```

### age_immigration
'Birth': 4, 'Younger than 5 years old': 3, '5 to 10 years old': 2,
       'Older than 10 years old':1

```python
def age_imm(row):
    if row['age_immigration'] == 'Birth':
        return 4
    if row['age_immigration'] == 'Younger than 5 years old':
        return 3
    if row['age_immigration'] == '5 to 10 years old': 
        return 2
    if row['age_immigration'] == 'Older than 10 years old':
        return 1
    
```

```python
df_fil_scores['qage_immigration'] = df_fil_scores.apply(lambda row : age_imm(row), axis = 1)
```

```python
df_fil_scores
```

```python
df_fil_scores.info()
```

```python
df_fil_scores['sex'].unique()
```

```python
def gender(row):
    if row['sex'] == 'Male':
        return 1
    if row['sex']=='Female':
        return 0
    
```

```python
df_fil_scores['qsex'] = df_fil_scores.apply(lambda row : gender(row), axis = 1)
```

```python
df_fil_scores.to_csv('scores_quantified.csv')
```
